## 1. Math operators and expressions

Python evaluates expressions using operator precedence. From **highest** to **lowest** precedence:

| Operators | Operation        | Example         |
| --------- | ---------------- | --------------- |
| `**`      | Exponent         | `2 ** 3 = 8`    |
| `%`       | Modulus/Remainder| `22 % 8 = 6`    |
| `//`      | Integer division | `22 // 8 = 2`   |
| `/`       | Division         | `22 / 8 = 2.75` |
| `*`       | Multiplication   | `3 * 3 = 9`     |
| `-`       | Subtraction      | `5 - 2 = 3`     |
| `+`       | Addition         | `2 + 2 = 4`     |

Use parentheses to change the order of evaluation.


```python
# Examples
print(2 + 3 * 6)      # 3*6 first, then +2 -> 20
print((2 + 3) * 6)    # 5*6 -> 30
print(2 ** 8)         # 256
print(23 // 7)        # integer division -> 3
print(23 % 7)         # remainder -> 2
print((5 - 1) * ((7 + 1) / (3 - 1)))  # 16.0
```

## 2. Data types

| Type   | Description           | Examples |

### Strings: concatenation and replication

- **Concatenation:** `+` joins two strings.
- **Replication:** `*` with an integer repeats the string.


```python
print('Alice' + 'Bob')
print('Hi! ' * 3)
s = 'Alice ' * 2 + 'Gupta'
print(s)
print(len(s))  # number of characters
```

## 3. Variables and naming

Rules for variable names:
1. One word (use `_` for multi-word: `user_name`).
2. Only letters, numbers, and underscore.
3. Cannot start with a number.

**Why this matters for AI:** We use variables for API keys (from env), base URLs, and payloads. Clear names make API code easier to read and debug.


```python
count = 10
api_base = "https://api.example.com"
print(count, api_base)
# count and api_base hold values we can reuse
```

## 4. Comments and docstrings

- **Single-line comment:** `#` to the end of the line.
- **Docstring:** Triple-quoted string right after `def` or at the top of a module; used by `help()`.


```python
# This is a comment; the next line is code
value = 1  # inline comment

def greet(name):
    """Return a greeting string. This is the function docstring."""
    return f"Hello, {name}"

print(greet("World"))
# help(greet)  # uncomment to see docstring in help()
```

## 5. The print() function

`print()` writes to the console. You can pass multiple values; by default they are separated by a space and end with a newline.

**f-strings** (Python 3.6+): Use `f"text {expression}"` to embed values in strings.


```python
print('Hello world!')
a, b = 1, 2
print('a =', a, 'b =', b)
print(f'Sum: {a} + {b} = {a + b}')
print('Line 1', end=' '); print('Line 2')  # end=' ' suppresses newline
```

## 6. Type conversion: str(), int(), float()

Convert between types when reading input or building API payloads.


```python
print(str(29))
print(str(-3.14))
print(int(7.7))
print(int(7.7) + 1)
# float to int truncates toward zero
age = 29
print('I am ' + str(age) + ' years old.')
print(f'I am {age} years old.')
```

## 7. Flow control – Comparison and Boolean operators

**Comparison operators:** `==`, `!=`, `<`, `>`, `<=`, `>=`. They evaluate to `True` or `False`.

**Boolean operators:** `and`, `or`, `not`. Use them to combine conditions.


```python
print(42 == 42)
print(40 == 42)
print('hello' == 'Hello')
print('dog' != 'cat')
print(42 == 42.0)
print(42 == '42')  # different types

print((4 < 5) and (5 < 6))
print((1 == 2) or (2 == 2))
print(not (1 == 2))
```

## 8. if / elif / else

Conditional execution. Only one branch runs. Indentation (typically 4 spaces) defines the block.


```python
name = 'Alice'
if name == 'Alice':
    print('Hi, Alice.')
elif name == 'Bob':
    print('Hi, Bob.')
else:
    print('Hello, stranger.')

# Another example
age = 25
if age < 12:
    print('kid')
elif age < 18:
    print('teen')
else:
    print('adult')
```

## 9. while and for loops

- **while:** Repeats while a condition is true.
- **for:** Iterates over a sequence (list, range, etc.).
- **break:** Exit the loop immediately.
- **continue:** Skip to the next iteration.


```python
# while
n = 0
while n < 3:
    print('n =', n)
    n = n + 1

# for with range(start, stop, step). stop is exclusive
print('range(5):', list(range(5)))
print('range(2, 10, 2):', list(range(2, 10, 2)))
for i in range(3):
    print('i =', i)
```

```python
# for over a list
fruits = ['apple', 'banana', 'cherry']
for f in fruits:
    print(f)

# break and continue (example)
for i in range(5):
    if i == 2:
        continue
    if i == 4:
        break
    print(i)
```

## 10. Lists

Lists are **ordered, mutable** sequences. They are zero-indexed. Negative indices count from the end (`-1` is last).

**Why this matters for AI:** Chat APIs expect a **list of message dicts** (e.g. `[{"role": "user", "content": "..."}]`). Lists are also used for batch payloads and results.


```python
spam = ['cat', 'bat', 'rat', 'elephant']
print(spam[0], spam[2])
print(spam[-1], spam[-2])
print(spam[1:3])
print(spam[:2])
print(spam[1:])
print(spam[::-1])
print(len(spam))
```

```python
# Lists are mutable
spam = ['cat', 'bat', 'rat']
spam[1] = 'aardvark'
print(spam)
spam.append('dog')
print(spam)
# Concatenation and replication
print([1, 2] + [3, 4])
print([0] * 3)
```

## 11. Tuples

Tuples are **ordered, immutable** sequences. Use parentheses. Often used for fixed pairs or return values.


```python
eggs = ('hello', 42, 0.5)
print(eggs[0])
print(eggs[1:3])
# eggs[0] = 'x'  # would raise TypeError
t = (1, 2)
a, b = t
print(a, b)
```

## 12. Dictionaries

Dictionaries store **key–value pairs**. Keys are unique; values can be any type.

**Why this matters for AI:** API request and response bodies are often represented as dicts (e.g. `{"model": "gpt-4o-mini", "messages": [...]}`). We build and parse these with dicts and `json`.


```python
my_cat = {'size': 'fat', 'color': 'gray', 'disposition': 'loud'}
print(my_cat['size'])
print(my_cat.get('color'))
print(my_cat.get('missing', 'default'))

# keys(), values(), items()
print(list(my_cat.keys()))
print(list(my_cat.values()))
for k, v in my_cat.items():
    print(k, '->', v)
```

## 13. Functions

Functions are defined with `def`. Parameters can have default values. Use `return` to send a value back.

- **Positional arguments:** Matched by order.
- **Keyword arguments:** Matched by name; can follow positional.
- **Default arguments:** Must come after non-default parameters.
- **`*args`:** Collects extra positional args as a tuple.
- **`**kwargs`:** Collects extra keyword args as a dict.


```python
def greet(name, greeting='Hello'):
    """Return a greeting string."""
    return f'{greeting}, {name}!'

print(greet('Alice'))
print(greet('Bob', greeting='Hi'))

def add(a, b):
    return a + b
print(add(3, 5))
```

```python
# Default arguments: non-default must come first
def describe(name, age=20):
    return f'{name} is {age} years old'
print(describe('Alice'))
print(describe('Bob', 25))

# *args and **kwargs (useful for flexible APIs)
def log_args(*args, **kwargs):
    print('args:', args)
    print('kwargs:', kwargs)
log_args(1, 2, x=10, y=20)
```

## 14. Scope: local vs global

Variables assigned inside a function are **local** unless declared `global`. Reading a name looks in local scope first, then global.


```python
x = 100
def f():
    y = 200
    print('inside f: x =', x, 'y =', y)
f()
# print(y)  # NameError: y is local to f
print('outside: x =', x)
```
